```python
#import modules and data
from	sklearn.datasets	import load_breast_cancer
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

breast_cancer = load_breast_cancer()

X=pd.DataFrame(breast_cancer.data, columns= breast_cancer.feature_names)
Y=breast_cancer.target
```

```python
# Split dataset into training and test sets
'''Para decidir los tamaños de los datos de test y los datos de training:
      - El tamaño de training data debe ser bastante grande para que al entrenar el modelo obtengamos las predicciones más exactas posibles
      - Para el modelo knn: Dimensionalidad de los datos: 30  -> curse of dimensionality: cuantas más dimensiones hay, más dificil es encontrar vecinos cerca. 
        Para que pueda haber una mayor cantidad de vecinos, se debe aumentar aún más el tamaño del training data.
   
   Por lo tanto:          training -> 80 % de los datos                  testing  -> 20 % de los datos'''

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, train_size = 0.8, random_state=0)
```

```python
#KNN MODEL
#chose the number of neighbors

'''Para decidir el número de vecinos (k), usaremos la regla heurística k = sqrt(n), y a partir de ahí iremos comprobando si hay algun valor de k cercano
   a dicho valor que de mejores resultados'''

k = 9   # mejor resultado que con k = sqrt(n) = 5
model_knn =KNeighborsClassifier(k)
```

```python

#Evaluate the Model

#Fit the classifier on the Training Set
model_knn.fit(X_train,Y_train)

#Make the predictions on the Test set
Y_pred_train = model_knn.predict(X_train)
Y_pred_test = model_knn.predict(X_test)

'''Para que el modelo esté bien hecho, la precisión al introducir en el modelo los datos de training y la precisión
   al introducir en el modelo los datos de test debe ser parecida, si es muy diferente habría que ajustar el modelo.'''

print("Percentage of correct predictions (training): %.3f" %(100*metrics.accuracy_score(Y_train,Y_pred_train)), "%")
print("Percentage of correct predictions (test)    : %.3f" %(100*metrics.accuracy_score(Y_test,Y_pred_test)), "%")

print('MSE: %.2f'% mean_squared_error(Y_test, Y_pred_test))
```

```python
#MODELO DE ÁRBOLES

'''El parámetro max_depth indica la profundidad máxima que puede tener el árbol. Reduciendo el valor de max_depth permitimos 
   que no se produzca un sobreajuste en el árbol, puesto que éste no se desarrollará lo suficiente como para aprender errores inherentes al training data'''
model_tree =  DecisionTreeClassifier(max_depth = 3, random_state=0)
model_tree.fit(X_train, Y_train)

```

```python
#Evaluate the Model

#Fit the classifier on the Training Set
Y_pred_train= model_tree.predict(X_train)

#Make the predictions on the Test set
Y_pred_train = model_tree.predict(X_train)
Y_pred_test = model_tree.predict(X_test)


'''Para que el modelo esté bien hecho, la precisión al introducir en el modelo los datos de training y la precisión
   al introducir en el modelo los datos de test debe ser parecida, si es muy diferente habría que ajustar el modelo.'''

print("Percentage of correct predictions (training): %.3f" %(100*metrics.accuracy_score(Y_train,Y_pred_train)), "%")
print("Percentage of correct predictions (test)    : %.3f" %(100*metrics.accuracy_score(Y_test,Y_pred_test)), "%")

print('MSE: %.2f'% mean_squared_error(Y_test, Y_pred_test))
```

```python
#MODELO DE BAYES

#fit model
'''Para estimar los valores de la distribución prior no contamos con el conocimiento suficiente como para estimar de forma razonada las probabilidades de que un tumor sea maligno o benigno, 
   por lo que usaremos priors no informativos, es decir, se estimará la misma probabilidad para ambas opciones, de modo que la distribución prior sea una distribución uniforme'''
model_nb=GaussianNB(priors = [0.5, 0.5])
model_nb.fit(X_train, Y_train)
```

```python
#Evaluate the Model

#Fit the classifier on the Training Set
Y_pred_train= model_nb.predict(X_train)

#Make the predictions on the Test set
Y_pred_train = model_nb.predict(X_train)
Y_pred_test = model_nb.predict(X_test)


print("Percentage of correct predictions (training): %.3f" %(100*metrics.accuracy_score(Y_train,Y_pred_train)), "%")
print("Percentage of correct predictions (test)    : %.3f" %(100*metrics.accuracy_score(Y_test,Y_pred_test)), "%")

print('MSE: %.2f'% mean_squared_error(Y_test, Y_pred_test))
```

```python
#leer los datos del archivo diabetes.csv
datos= pd.read_csv ('diabetes.csv')
X = datos.iloc[:,:-1]
Y = datos.iloc[:,-1]
```

```python
#KNN
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.2, random_state=0)
#chose the number of neighbors

k = 3   # k = sqrt(n)
model_knn =KNeighborsClassifier(k)

#Fit the classifier on the Training Set
model_knn.fit(X_train,Y_train)

#Evaluate the Model

#Make the predictions on the Test set
Y_pred_train = model_knn.predict(X_train)
Y_pred_test = model_knn.predict(X_test)

print("Percentage of correct predictions (training): %.3f" %(100*metrics.accuracy_score(Y_train,Y_pred_train)), "%")
print("Percentage of correct predictions (test)    : %.3f" %(100*metrics.accuracy_score(Y_test,Y_pred_test)), "%")
print('MSE: %.2f' % mean_squared_error(Y_test, Y_pred_test))
```

```python
#Arboles
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)

#fit model
model_tree =  DecisionTreeClassifier(max_depth = 10, random_state=0)
model_tree.fit(X_train, Y_train)


#Fit the classifier on the Training Set
Y_pred_train= model_tree.predict(X_train)

#Make the predictions on the Test set
Y_pred_test= model_tree.predict(X_test)

print("Percentage of correct predictions (training): %.3f" %(100*metrics.accuracy_score(Y_train,Y_pred_train)), "%")
print("Percentage of correct predictions (test)    : %.3f" %(100*metrics.accuracy_score(Y_test,Y_pred_test)), "%")


print('MSE: %.2f'% mean_squared_error(Y_test, Y_pred_test))
```

```python
#Naive Bayes
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)

#fit model
model_nb=GaussianNB(priors = [0.5, 0.5])
model_nb.fit(X_train, Y_train)

#prediction
Y_pred_train=model_nb.predict(X_train)
Y_pred_test=model_nb.predict(X_test)

print("Percentage of correct predictions (training): %.3f" %(100*metrics.accuracy_score(Y_train,Y_pred_train)), "%")
print("Percentage of correct predictions (test)    : %.3f" %(100*metrics.accuracy_score(Y_test,Y_pred_test)), "%")

print('MSE: %.2f'% mean_squared_error(Y_test, Y_pred_test))
```
